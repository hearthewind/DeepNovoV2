```python
import pickle
with open('/home/m/data/DeepNovoV2/westlake_train_data_with_acetylation/db.psms.backup.csv', 'r') as fr:
    with open('/home/m/data/DeepNovoV2/westlake_train_data_with_acetylation/spectrum_location.pytorch.pkl', 'rb') as fs:
        with open('/home/m/data/DeepNovoV2/westlake_train_data_with_acetylation/db.psms.csv', 'w') as fw:
            spectrum_location_dict = pickle.load(fs)
            line = fr.readline()
            fw.write(line)
            while line:
                line = fr.readline()
                
                try:
                    line_tokens = line.split(',')
                    file_name = line_tokens[11].replace('"', '') + '.refined.mgf'
                    scan = str(line_tokens[10])
                    
                except IndexError:
                    continue
                
                try:
                    _ = spectrum_location_dict[(file_name, scan)]
                    fw.write(line)
                except KeyError:
                    pass
```

```python
import csv
scan_seq_map = {}
with open('/home/m/data/DeepNovoV2/new_data/db.psms.csv', 'r') as fr:
    reader = csv.reader(fr, delimiter=',')
    for line in reader:
        seq = line[0]
        score = line[1]
        mass = line[2]
        file = line[11]
        scan = line[10]
        
        try:
            old_seq, old_mass, old_score = scan_seq_map[(file, scan)]
            if old_seq != seq:
                print('file: {}'.format(file))
                print('scan: {}'.format(scan))
                print('old_seq: {}'.format(old_seq))
                print('new_seq: {}'.format(seq))
                print('old_mass: {}'.format(old_mass))
                print('new_mass: {}'.format(mass))
                print('old_score: {}'.format(old_score))
                print('new_score: {}'.format(score))
                print('---------------')
        except KeyError:
            scan_seq_map[(file, scan)] = (seq, mass, score)
```
